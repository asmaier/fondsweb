```python
from lxml import html
# see https://stackoverflow.com/questions/49087990/python-request-being-blocked-by-cloudflare
import cloudscraper
from fondsweb import fondsweb as fw
```

```python
scraper = cloudscraper.create_scraper()
```

```python
url = "https://www.fondsweb.com/de/suchen"
```

```python
response = scraper.get(url)
```

```python
response
```

```python
tree = html.fromstring(response.content)
```

```python
selections = {}
for elem in tree.xpath("//select[@name]"):

    name = elem.name
    
    values = tree.xpath(f"//select[@name='{name}']/option/@value")
    texts = tree.xpath(f"//select[@name='{name}']/option/text()")
    # remove the the number in brackets from the end of each string 
    texts = [text[0:text.rfind("(")].strip() for text in texts]

    selections[name] = dict(zip(texts, values))
selections    
```

```python
values = tree.xpath("//input[@name='yield']/@value")
texts = tree.xpath("//input[@name='yield']/../label/text()")
yield_options = dict(zip(texts, values))
yield_options
```

```python
values = tree.xpath("//input[@name='etf']/@value")
texts = tree.xpath("//input[@name='etf']/../../td/label/text()")
etf_options = dict(zip(texts, values))
etf_options
```

```python
values = tree.xpath("//select[@name='sort']/option/@value")
texts = tree.xpath("//select[@name='sort']/option/text()")
sort_options = dict(zip(texts, values))
sort_options
```
