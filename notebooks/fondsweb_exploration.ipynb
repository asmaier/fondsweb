```python
import pandas as pd
import plotly.express as px

from fondsweb import fondsweb as fw
```

```python
fw.config.sort_options
```

```python
options1j = {'PremiumService-Fonds': 'psf',
 'Fondsname': 'fundname',
 'KVG-Name': 'issuername',
 'Performance laufendes Jahr': 'perytd',
 'Performance 1 Jahr': 'per1j',
 'Performance 3 Jahre': 'per3j',
 'Performance 5 Jahre': 'per5j',
 'Performance 10 Jahre': 'per10j',
 'Volatilität 1 Jahr': 'vol1j',
# 'Volatilität 3 Jahre': 'vol3j',
# 'Volatilität 5 Jahre': 'vol5j',
# 'Volatilität 10 Jahre': 'vol10j',
 'Sharpe Ratio 1 Jahr': 'sharpe1j',
# 'Sharpe Ratio 3 Jahre': 'sharpe3j',
# 'Sharpe Ratio 5 Jahre': 'sharpe5j',
 'FWW FundStars®': 'fundstars',
 'ISS ESG Fund Rating': 'esgrating',
 'SRI': 'sri',
 'Summe laufende Kosten': 'oc'}
```

```python
results1j = []
for option in options1j:
    res = fw.search_fondsweb(afocus=["Aktienfonds"], sort=option)
    results1j.extend(res)
```

```python
df_results = pd.DataFrame(results1j)
```

```python
# This should be true, if we used the consistent sort options. 
len(df_results.drop_duplicates(subset=["isin"])) == len(df_results.drop_duplicates())
```

```python
df_results = df_results.drop_duplicates()
```

```python
df_results = df_results.dropna(subset=["costs"])
```

```python
df_results["costs_inv"] = 1.0 / df_results["costs"]
```

```python
px.scatter(df_results, x="sharpe", y="current_year", hover_data=["name", "isin"], color="volat", size="costs_inv")
```

```python
# In analogy to the Risiko-Rendite Chart of f
px.scatter(df_results, x="volat", y="year_1", hover_data=["name", "isin"], color="sharpe", size="costs_inv")
```

```python
config = fw.fetch_config()
```

```python
funds={}
for index, row in df_results.iterrows():
    isin = row['isin']
    df = fw.get_fund_timeseries(isin, config=config)
    funds[isin] = df
```

```python
df_results['custom_date'] = None
for index, row in df_results.iterrows():
    isin = row['isin']
    df = funds[isin]
    perf = fw.get_fund_performance(df, "20250601")
    print(f"{isin}: {perf}")
    df_results.at[index, 'custom_date'] = perf
```

```python
px.scatter(df_results, x="current_year", y="custom_date", hover_data=["name", "isin"], color="volat", size="costs_inv")
```
